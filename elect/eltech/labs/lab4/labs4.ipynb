```python
import cmath as c
from math import *
import numpy as np
import pandas as pd
```

```python
def rowGen(data:complex):
	return (str(data), f"{abs(data)}e^({degrees(c.phase(data))}j)")
```

# 1.1
## Calculations

```python
data = np.zeros((7,4),dtype=object)
U = complex(10,0)
Zc = complex(0, -(1/(2*pi*50*200*10**-6)))
Zr = complex(0.01,0)
Z = Zr + Zc
I = U/Z
Uc = I*Zc
Ur = I*Zr
Ssrc = I*U
Spr = (I**2)*0.01+(I**2)*Zc
```

```python
data[0,0],data[0,1] = rowGen(U)
data[1,0],data[1,1] = rowGen(Zc)
data[2,0],data[2,1] = rowGen(I)
data[3,0],data[3,1] = rowGen(Uc)
data[4,0],data[4,1] = rowGen(Ur)
data[5,0],data[5,1] = rowGen(Ssrc)
data[6,0],data[6,1] = rowGen(Spr)
```

## Measurements

```python
U = c.rect(10,0)
I =  c.rect(0.63,pi/2)
tau = 0.01*200*10**-6
phi = tau*2*pi*50
phi = degrees(atan(phi))
Uc = c.rect(14/(2**.5),radians(0-phi))
Ur = c.rect((9.9*10**-3)/(2**.5),radians(90-phi))
Zc = Uc / I
Zr = Ur / I
Ssrc = I*U
Spr = (I**2)*0.01+(I**2)*Zc
```

```python
data[0,2],data[0,3] = rowGen(U)
data[1,2],data[1,3] = rowGen(Zc)
data[2,2],data[2,3] = rowGen(I)
data[3,2],data[3,3] = rowGen(Uc)
data[4,2],data[4,3] = rowGen(Ur)
data[5,2],data[5,3] = rowGen(Ssrc)
data[6,2],data[6,3] = rowGen(Spr)
```

```python
pd.DataFrame(data=data,index=["E","Zc","I","Uc","Ur","Ssrc","Spr"], columns=["rect_calc","exp_calc","rect_mes","exp_mes"])
```

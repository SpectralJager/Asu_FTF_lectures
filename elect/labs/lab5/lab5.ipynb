```python
import numpy as np
import cmath as c
import pandas as pd
from math import *
```

```python
def rowGen(data:complex):
	return (str(data), f"{abs(data)}e^({degrees(c.phase(data))}j)")
```

## Calculations

```python
data = np.zeros((12,4),dtype=object)
E = complex(10,0)
f = 50
R1 = complex(10,0)
R2 = complex(8,0)
R3 = complex(6,0)
C3 = 400*10**-6
L2 = 60*10**-3
omega = 2*pi*f
```

```python
Zc3 = complex(0,-(1/(omega*C3)))
Zcd = Zc3 + R3
Zl2 = complex(0,omega*L2)
Zab = Zl2 + R2
Zabcd = (1/Zcd+1/Zab)**-1
Z = Zabcd + R1
I = E/Z
Ur1 = I*R1
Uab = I*Zabcd
Iab = Uab/Zab
Ur2 = Iab * R2
Ul2 = Iab * Zl2
Icd = Uab/Zcd
Uc3 = Icd*Zc3
Ur3 = Icd*R3
```

```python
Ssrs = I*E
Spr = I*Ur1+Iab*Uab+Icd*Uab
```

```python
data[0,0],data[0,1] = rowGen(E)
data[1,0],data[1,1] = rowGen(Z)
data[2,0],data[2,1] = rowGen(I)
data[3,0],data[3,1] = rowGen(Iab)
data[4,0],data[4,1] = rowGen(Icd)
data[5,0],data[5,1] = rowGen(Ur1)
data[6,0],data[6,1] = rowGen(Ur2)
data[7,0],data[7,1] = rowGen(Ur3)
data[8,0],data[8,1] = rowGen(Ul2)
data[9,0],data[9,1] = rowGen(Uc3)
data[10,0],data[10,1] = rowGen(Ssrs)
data[11,0],data[11,1] = rowGen(Spr)
```

## Measurements

```python
phi = f*(625*10**-6)*2*pi
I = c.rect(0.477,phi)
Uab = E-Ur1
Iab = Uab/Zab
Icd = Uab/Zcd
Ur2 = Iab*R2
Ul2 = Iab*Zl2
Ur3 = Icd*R3
Uc3 = Icd*Zc3
Ssrs = I*E
Spr = I*Ur1+Iab*Uab+Icd*Uab
```

```python
data[0,2],data[0,3] = rowGen(E)
data[1,2],data[1,3] = rowGen(Z)
data[2,2],data[2,3] = rowGen(I)
data[3,2],data[3,3] = rowGen(Iab)
data[4,2],data[4,3] = rowGen(Icd)
data[5,2],data[5,3] = rowGen(Ur1)
data[6,2],data[6,3] = rowGen(Ur2)
data[7,2],data[7,3] = rowGen(Ur3)
data[8,2],data[8,3] = rowGen(Ul2)
data[9,2],data[9,3] = rowGen(Uc3)
data[10,2],data[10,3] = rowGen(Ssrs)
data[11,2],data[11,3] = rowGen(Spr)
```

```python
pd.DataFrame(data=data,index=["E","Z","I","Iab","Icd","Ur1","Ur2","Ur3","Ul2","Uc3","Ssrc","Spr"], columns=["rect_calc","exp_calc","rect_mes","exp_mes"])
```

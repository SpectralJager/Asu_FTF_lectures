```python
%matplotlib inline
#--------------------
import numpy as np
import matplotlib.pyplot as plt
```

Первое задание

```python
n = 30

g = lambda t,f: np.cos(2*np.pi*f*t)
f = 35
T = 1/f

fd = 100
Td = 1/fd
fm = fd - f

t = np.arange(0,Td*n,Td)
t1 = np.arange(0,Td*n,Td*0.0001)


plt.figure(figsize=(20,10))
plt.title('cos(2*pi*f*t)') 
plt.plot(t1,g(t1,f),label=f'f = {f}',color='red')
plt.plot(t1,g(t1,fm),label=f'fm = {fm}',color='green')
plt.stem(t,g(t,f),label=f'fd = {fd}')
plt.xlim([0, t[-1]])
#plt.grid(True)
plt.legend()
plt.show()

print(f'T = {round(T,4)}, f = {f}')
print(f'Tm = {round(1/fm,4)}, fm = {fm}')
print(f'Td = {round(Td,4)}, fd = {fd}')
        
```

Второе задание

```python
print(f'T = {round(T,4)}, f = {f}')
print(f'Tm = {round(1/fm,4)}, fm = {fm}')
print(f'Td = {round(Td,4)}, fd = {fd}')
fd += 50
Td = 1/fd
print(f'new fd = {fd}, new Td = {round(Td,4)}')

t = np.arange(0,Td*n,Td)
t1 = np.arange(0,Td*n,Td*0.0001)

plt.figure(figsize=(20,10))
plt.title('cos(2*pi*f*t)') 
plt.plot(t1,g(t1,f),label=f'f = {f}',color='red')
plt.plot(t1,g(t1,fm),label=f'fm = {fm}',color='green')
plt.stem(t,g(t,f),label=f'fd = {fd}')
plt.xlim([0, t[-1]])
#plt.grid(True)
plt.legend()
plt.show()
```

Третье задание

```python
fs = 70
fd = 150
Td = 1/fd
print(f'fs = {fs}, fd = {fd}, T = {round(T,4)}')

h = lambda t, f: np.cos(2*np.pi*t*f)


t = np.arange(0,Td*n,Td)
t1 = np.linspace(0,Td*n,1000)

y = []

for j in range(1000):
    s = 0
    for i in range(n):
        temp = (np.pi*t1[j])/Td - i*np.pi
        s += h(t[i],fs) * (1 if temp == 0 else np.sin(temp)/temp)
    y.append(s)



plt.figure(figsize=(20,10))
plt.stem(t,h(t,fs))
#plt.plot(t1,h(t1,fs))
plt.plot(t1,y)
plt.ylim(-2,2)
plt.xlim(0,t[-1])
plt.show()
```

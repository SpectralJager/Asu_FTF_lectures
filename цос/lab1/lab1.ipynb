```python
#%matplotlib inline
#--------------------
import numpy as np
import matplotlib.pyplot as plt
```

Первое задание

```python
n = 30

g = lambda t,f: np.cos(2*np.pi*f*t)
f = 35
T = 1/f

fd = 100
Td = 1/fd
fm = fd - f

t = np.arange(0,Td*n,Td)
t1 = np.arange(0,Td*n,Td*0.0001)


plt.figure(figsize=(20,10))
plt.title('cos(2*pi*f*t)') 
plt.plot(t1,g(t1,f),label=f'f = {f}',color='red')
plt.plot(t1,g(t1,fm),label=f'fm = {fm}',color='green')
plt.stem(t,g(t,f),label=f'fd = {fd}')
plt.xlim([0, t[-1]])
#plt.grid(True)
plt.legend()
plt.show()

print(f'T = {round(T,4)}, f = {f}')
print(f'Tm = {round(1/fm,4)}, fm = {fm}')
print(f'Td = {round(Td,4)}, fd = {fd}')
        
```

Второе задание

```python
print(f'T = {round(T,4)}, f = {f}')
print(f'Tm = {round(1/fm,4)}, fm = {fm}')
print(f'Td = {round(Td,4)}, fd = {fd}')
fd += 50
Td = 1/fd
print(f'new fd = {fd}, new Td = {round(Td,4)}')

t = np.arange(0,Td*n,Td)
t1 = np.arange(0,Td*n,Td*0.0001)

plt.figure(figsize=(20,10))
plt.title('cos(2*pi*f*t)') 
plt.plot(t1,g(t1,f),label=f'f = {f}',color='red')
plt.plot(t1,g(t1,fm),label=f'fm = {fm}',color='green')
plt.stem(t,g(t,f),label=f'fd = {fd}')
plt.xlim([0, t[-1]])
#plt.grid(True)
plt.legend()
plt.show()
```

Третье задание

```python
def H(t, n: int, Td : float, h): # Kot
    s = 0
    for i in range(n):
        temp = (np.pi*t)/Td - i*np.pi
        s += h(Td*i) * (1 if temp == 0 else np.sin(temp)/temp)
    return s

def h(t): # func
    # fs = 70
    # return np.cos(2*np.pi*fs*t)
    if t < 0:
        return -0.5
    elif t < 2:
        return np.sqrt(4 * t - t**2)
    elif t < 4:
        return 6 - 2 * t
    elif t < 6:
        return -np.sqrt(-t * t + 8 * t - 12)
    else:
        return -0.5

# Discr Sig
t_start = -2
N = 30
# fd = 150
# Td = 1/fd
Td = (8+2)/N

# t1 = np.arange(-2,N*Td,Td)
t1 = np.linspace(t_start,8,N)
h_res = []
for t in t1:
    h_res.append(h(t))

# Rec sig

t2 = np.linspace(t_start,(N-1)*Td,1000)
H_res = []
for t in t2:
    H_res.append(H(t,100,Td,h))

plt.figure(figsize=(20,10), dpi=300)

plt.stem(t1,h_res)
plt.plot(t2,H_res)

plt.show()
```

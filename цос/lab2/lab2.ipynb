```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift
import random
import os
```

```python
# 1
fd = 150
Td = 1/fd
n=5
f = list(np.linspace(0,fd,n+1))
f.pop(-1)
A = [round(random.random(),3) for _ in range(len(f))]
g = lambda a,t,f: a*np.sin(2*np.pi*f*t)
t = np.arange(0,0.1,Td)
y = []
for f_i,a in zip(f,A):
    y.append(g(a,t,f_i))
y = np.array(y)
y = y.sum(axis=0)

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.stem(t,y)
plt.xlim(0,t[-1])

lfft = list(fft(y))
lfft = [abs(i) for i in lfft]
lfft.append(f[0])
f_range = np.linspace(-fd/2,fd/2,len(lfft))
plt.subplot(122)
plt.bar(f_range,lfft,width=fd/len(lfft))
plt.show()
```

```python
f1 = f[random.randint(0,n-1)]+1
a1 = round(random.random(),3)
lfft = list(fft(g(a1,t,f1)))
lfft = [abs(i) for i in lfft]
lfft.append(f[0])
plt.figure(figsize=(20,10))
plt.subplot(111)
plt.bar(f_range,lfft,width=fd/len(lfft))
plt.show()
```

```python


for k in range(1,28):
    with open(os.path.join(os.path.abspath(os.getcwd()),'signals',f'signal_{k}.dat'),'r') as f:
        temp = [float(data[0:-1]) for data in f]
        fd = 1
        Td = 1/fd
        lfft = list(fft(temp))
        lfft = [abs(i) for i in lfft]
        lfft.append(lfft[0])
        f_range = np.linspace(-fd/2,fd/2,len(lfft))
        plt.bar(f_range,lfft,width=fd/len(lfft)*.7)
        lim_fft = 5
        f = []
        a0 = 0
        a = []
        Y = fftshift(fft(temp))
        p = np.angle(Y)
        p[np.abs(Y) < 1] = 0
        phase = [round(p[i],3) for i in range(len(p)//2,len(p)) if p[i] != 0.0]
        for i in range(len(lfft)):
            if lfft[i] > lim_fft:
                plt.annotate(round(f_range[i],3),(f_range[i],lfft[i]))
                temp = round(abs(f_range[i]),3)
                if f.count(temp) == 0:
                    phase.append(p[i])
                    a.append(2*round(lfft[i],3))
                    f.append(temp)
                if f_range[i] == 0:
                    a0 = round(lfft[i],3)
        print(f'\nf = {f}, \na0 = {a0}, a = {a}, \nphase = {np.degrees(phase[:len(phase)//2])}\n')
        plt.xlim(0)
        plt.show()
```

```python
n = len(y)
X_avr = np.sum(x)/n
Y_avr = np.sum(y)/n
Sx = (x-X_avr)
Sy = (y-Y_avr)
```

```python
r = np.sum((Sx)*(Sy))/np.sqrt(np.sum(Sx**2) * np.sum(Sy**2))
```

```python
mr = (1-r**2)/np.sqrt(n-1)
t = r/mr
```

```python
data_pd = pd.DataFrame()
data_pd['X'] = x
data_pd['Y'] = y
```

```python
data_pd2 = pd.DataFrame()
data_pd2['X_avr'] = ['Y_avr','r','mr','t']
data_pd2[f'{X_avr}'] = [Y_avr,r,mr,t]
```

```python
data_pd = pd.concat([data_pd,data_pd2],axis=1)
```
